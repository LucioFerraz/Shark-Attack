<h4>Sétimo passo: criação e aplicação de uma função def ao DataFrame. </h4>

<p> Utilizou-se a função list, para elencar todas as espécies de tubarões presentes do dataset, criação e aplicação de uma função ao DataFrame, responsável por buscar argumentos iguais dentre as espécies de tubarões para criação de uma nova coluna.<p>

```python
'''Com a função list, foram trazidos em grupos únicos todas as espécies de tubarões elencadas nos registros de ataque.''' 

list(df['Species '].unique())[:5]
```

```python
'''Criação de uma função (chamada tubarão_branco) no qual o argumento white shark buscará argumentos iguais dentre as espécies, 
para tanto utilizou-se o .lower() para conferir correspondência. Caso não seja localizado, retorna o mesmo argumento x.'''

def tubarao_branco(x):
    try: 
        if 'white shark' in x.lower():
            return 'white shark'
        else: 
            return x 
    except: 
        return x
```

```python
'''Aplicação da função acima definada, por meio do .apply() e criação de uma coluna nomeada como white_shark, uma vez que 
conterá os white sharks como registros, independente da grafia que foi empregada anteriormente. Total 11 colunas. '''

df['white_shark']=df['Species '].apply(tubarao_branco)
df
```

```python
'''Função aplicada para que apenas os registros da espécie white shark, presente na coluna white_shark sejam chamados.'''

df[df['white_shark']=='white shark']
```

<h4>Oitavo passo: Agregação de colunas, contagem dos registros e análise dos resultados.</h4>

<p>Agregação de colunas por meio do groupby, contagem de registros de tubarão branco, organizados de forma decresente por meio de .count().sort_values(), renomeação da coluna por meio de .rename(columns={'x':'xx'}) e criação de novo index. Da análise destes resultados, serão conferidos os insights relativos a business question.<p>

```python
'''Agregação das colunas Country e Area, contagem dos registros de ataques de tubarões brancos, organizados de forma decrescente
renomeação de coluna e criação de index por meio da função reset_index().'''

df_Country = df.groupby(['Country', 'Area']).count().sort_values('white_shark', ascending=False)[['white_shark']].rename(columns={'white_shark':'shark_attack_counter'}).reset_index()
df_Country
```
